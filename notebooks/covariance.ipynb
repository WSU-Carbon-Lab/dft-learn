```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# read the cov matrix from the csv
cov = pd.read_csv("cov.csv")
names = pd.read_csv("names.csv", header=None, skiprows=1)
# drop the first row of the cov dataframe
cov = cov.drop(cov.index[0]).reset_index(drop=True)
cov
```

```python
# reneame the columns
cov.columns = names[0].values
# Create a new column for the index
cov["index"] = names[0].values
# set the index to the new column
cov.set_index("index", inplace=True)
cov
```

```python
# get numeric columns where the name ends in the pattern *_[0-9]
cov_numeric = cov.filter(regex=".*_[0-9]+$")
# slice the rows associated with this filer
cov_numeric = cov_numeric.loc[cov_numeric.index.str.contains("^.*_[0-9]+$")]
cov_energy = cov_numeric.filter(regex="^.*Energy_[0-9]+$")
cov_energy = cov_energy.loc[cov_energy.index.str.contains("^.*Energy_[0-9]+$")]
cov_width = cov_numeric.filter(regex="^.*Width_[0-9]+$")
cov_width = cov_width.loc[cov_width.index.str.contains("^.*Width_[0-9]+$")]
cov_amplitude = cov_numeric.filter(regex="^.*Amplitude_[0-9]+$")
cov_amplitude = cov_amplitude.loc[
    cov_amplitude.index.str.contains("^.*Amplitude_[0-9]+$")
]
cov_theta = cov_numeric.filter(regex="^.*Theta_[0-9]+$")
cov_theta = cov_theta.loc[cov_theta.index.str.contains("^.*Theta_[0-9]+$")]

display(cov_energy)
display(cov_width)
display(cov_amplitude)
display(cov_theta)
```

```python
cov_energy_matrix = cov_energy.values
cov_width_matrix = cov_width.values
cov_amplitude_matrix = cov_amplitude.values
cov_theta_matrix = cov_theta.values
# Create a mask for the upper triangle
mask = np.triu(np.ones_like(cov_energy_matrix, dtype=bool))
# set the diagonal to True
mask[np.diag_indices_from(mask)] = False


def plot_covariance_matrix(cov_matrix, cov_df, title, ax):
    sns.heatmap(
        cov_matrix,
        mask=mask,
        cmap="coolwarm",
        annot=True,
        fmt="0.1f",
        square=True,
        cbar_kws={"shrink": 0.8},
        xticklabels=cov_df.columns,
        yticklabels=cov_df.index,
        ax=ax,
    )
    ax.set_title(title)
    # split the columns at'_' and get the index from the end
    ax.set_xlabel("Peak Number")
    ax.set_ylabel("Peak Number")
    peak_labels = [col.split("_")[-1] for col in cov_df.columns]
    ax.set_xticklabels(peak_labels, rotation=45, ha="right")
    ax.set_yticklabels(peak_labels, rotation=0)


fig, axs = plt.subplots(2, 2, figsize=(18, 12))
plot_covariance_matrix(
    cov_energy_matrix, cov_energy, "Covariance Matrix of Energy", axs[0, 0]
)
plot_covariance_matrix(
    cov_width_matrix, cov_width, "Covariance Matrix of Width", axs[0, 1]
)
plot_covariance_matrix(
    cov_amplitude_matrix, cov_amplitude, "Covariance Matrix of Amplitude", axs[1, 0]
)
plot_covariance_matrix(
    cov_theta_matrix, cov_theta, "Covariance Matrix of Theta", axs[1, 1]
)
```

```python
cov_energy_matrix = cov_energy.values
cov_width_matrix = cov_width.values
cov_amplitude_matrix = cov_amplitude.values
cov_theta_matrix = cov_theta.values

# Create masks for the upper triangle
mask_energy = np.triu(np.ones_like(cov_energy_matrix, dtype=bool))
mask_width = np.triu(np.ones_like(cov_width_matrix, dtype=bool))
mask_amplitude = np.triu(np.ones_like(cov_amplitude_matrix, dtype=bool))
mask_theta = np.triu(np.ones_like(cov_theta_matrix, dtype=bool))

# Set diagonal to False to show it
mask_energy[np.diag_indices_from(mask_energy)] = False
mask_width[np.diag_indices_from(mask_width)] = False
mask_amplitude[np.diag_indices_from(mask_amplitude)] = False
mask_theta[np.diag_indices_from(mask_theta)] = False

# Calculate common color scale
all_values = np.concatenate(
    [
        cov_energy_matrix[~mask_energy].flatten(),
        cov_width_matrix[~mask_width].flatten(),
        cov_amplitude_matrix[~mask_amplitude].flatten(),
    ]
)
vmin, vmax = np.percentile(
    all_values, [5, 95]
)  # Use 5th-95th percentile for better visualization


def plot_covariance_matrix(cov_matrix, mask, title, ax, show_cbar=False):
    # Create heatmap without annotations for cleaner look
    im = ax.imshow(
        cov_matrix,
        cmap="RdBu_r",
        vmin=vmin,
        vmax=vmax,
        aspect="equal",
        interpolation="nearest",
    )

    # Apply mask
    masked_data = np.ma.masked_where(mask, cov_matrix)
    im = ax.imshow(
        masked_data,
        cmap="RdBu_r",
        vmin=vmin,
        vmax=vmax,
        aspect="equal",
        interpolation="nearest",
    )

    # Highlight diagonal (errors) with black outline
    for i in range(cov_matrix.shape[0]):
        rect = plt.Rectangle(
            (i - 0.5, i - 0.5), 1, 1, fill=False, edgecolor="black", linewidth=2
        )
        ax.add_patch(rect)

    ax.set_title(title, fontsize=14, pad=10)
    ticks = np.arange(cov_matrix.shape[0])
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.set_xticklabels([f"Peak {i + 1}" for i in ticks], rotation=45, ha="right")
    ax.set_yticklabels([f"Peak {i + 1}" for i in ticks])

    # Add error values as text on diagonal
    for i in range(cov_matrix.shape[0]):
        error = np.sqrt(cov_matrix[i, i])
        ax.text(
            i,
            i,
            f"{error:.2f}",
            ha="center",
            va="center",
            fontsize=12,
            fontweight="bold",
            color="white",
            bbox=dict(boxstyle="round,pad=0.2", facecolor="black", alpha=0.7),
        )

    return im


# Create figure with better spacing
fig, axs = plt.subplots(2, 2, figsize=(16, 12), sharey=True)

# Plot each matrix
im1 = plot_covariance_matrix(
    cov_energy_matrix, mask_energy, "Energy Covariance", axs[0, 0]
)
im2 = plot_covariance_matrix(
    cov_width_matrix, mask_width, "Width Covariance", axs[0, 1]
)
im3 = plot_covariance_matrix(
    cov_amplitude_matrix, mask_amplitude, "Amplitude Covariance", axs[1, 0]
)
im4 = plot_covariance_matrix(
    cov_theta_matrix, mask_theta, "Theta Covariance", axs[1, 1]
)

# Add a single colorbar for all plots
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cbar = fig.colorbar(im1, cax=cbar_ax)
cbar.set_label("Covariance", fontsize=12, rotation=270, labelpad=20)

# Add text annotations for errors
fig.text(
    0.5,
    0.75,
    "Black outlined cells show errors\n (diagonal elements)",
    ha="center",
    fontsize=12,
    style="italic",
)

plt.tight_layout()
plt.subplots_adjust(right=0.9)
plt.show()
```

```python
# gram the initial and final parameters
params_initial = params.filter(like="pwi0Fit")
params_final = params.filter(like="pwFinal")
params_final.to_csv("params_final.csv", index_label="Peak Number", index=False)
params_initial.to_csv("params_initial.csv", index_label="Peak Number", index=False)
```

```python
for peak in peaks:
    osx = params_numeric[f"OSxx_{peak}"].values
    osy = params_numeric[f"OSyy_{peak}"].values
    osz = params_numeric[f"OSzz_{peak}"].values
    adj = params_numeric[f"MolTensorTheta_{peak}"].values
    os_ip = np.rad2deg(np.sqrt(osx**2 + osy**2))  # calculate the in-plane component
    # calculate the polar angle from the in-plane component and the z component
    osTheta = np.arctan2(os_ip, osz)  # polar angle in radians
    params_numeric[f"tdmTheta_{peak}"] = np.rad2deg(osTheta) - adj  # convert to degrees
```

```python
params_energy = params_numeric.filter(regex="^.*Energy_[0-9]+$")
params_width = params_numeric.filter(regex="^.*Width_[0-9]+$")
params_amplitude = params_numeric.filter(regex="^.*Amplitude_[0-9]+$")
params_theta = params_numeric.filter(regex="^.*tdmTheta_[0-9]+$")


# plot parameter changes though each fitting step
def plot_parameter_changes(params_df, title, ax):
    params_df.plot(ax=ax, marker="o", legend=False, linestyle="-", lw=2)
    ax.set_xticks(range(len(params_df)))
    ax.set_xticklabels(
        [f"Step {i + 1}" for i in range(len(params_df))], rotation=45, ha="right"
    )
    ax.set_xlabel("Fitting Step")
    ax.set_ylabel(title)
    ax.grid(True)


fig, axs = plt.subplots(4, 1, figsize=(5, 12), sharex=True, sharey=False)
plot_parameter_changes(params_energy, "Energy Parameter Changes", axs[0])
plot_parameter_changes(params_width, "Width Parameter Changes", axs[1])
plot_parameter_changes(params_amplitude, "Amplitude Parameter Changes", axs[2])
plot_parameter_changes(params_theta, "Theta Parameter Changes", axs[3])
peaks = [col.split("_")[-1] for col in params_energy.columns]
plt.tight_layout()
plt.show()
```

```python
# Calculate the difference between the final and initial parameters
def calculate_parameter_differences(params_df):
    initial_params = params_df.iloc[0]
    final_params = params_df.iloc[-1]
    differences = final_params - initial_params
    return differences


# Calculate differences for each parameter set
energy_difference = calculate_parameter_differences(params_energy)
width_difference = calculate_parameter_differences(params_width)
amplitude_difference = calculate_parameter_differences(params_amplitude)
theta_difference = calculate_parameter_differences(params_theta)
# calc difference in number of standard deviations from the mean
energy_difference_std = energy_difference / errors_energy
width_difference_std = width_difference / errors_width
amplitude_difference_std = amplitude_difference / errors_amplitude
theta_difference_std = theta_difference
```

```python
fig, axs = plt.subplots(4, 1, figsize=(9, 12), sharex=True)
# plt.style.use('seaborn-v0_8-whitegrid') # Using a seaborn style for better aesthetics

peak_indices = np.arange(1, len(params_energy.columns) + 1)  # Peak numbers 1 to 17


def plot_barbell(ax, df, title):
    # Extract initial and final parameters
    initial_params = df.loc["pwi0Fit"]
    final_params = df.loc["pwFinal"]

    num_peaks = len(initial_params)
    peak_indices = np.arange(1, num_peaks + 1)  # Peak numbers 1 to 17

    # Plot vertical lines (stems of the barbell)
    for i in range(num_peaks):
        ax.plot(
            [peak_indices[i], peak_indices[i]],
            [initial_params[i], final_params[i]],
            marker="o",
            linestyle="-",
            color="gray",
            markersize=8,
            linewidth=1.5,
        )

    # Plot markers for initial and final values with different colors
    ax.scatter(
        peak_indices, initial_params, color="skyblue", s=100, zorder=3, label="Initial"
    )
    ax.scatter(
        peak_indices, final_params, color="tomato", s=100, zorder=3, label="Refined"
    )

    ax.set_ylabel(title, fontsize=12)
    ax.set_xticks(peak_indices)
    ax.tick_params(axis="x", rotation=0)
    ax.grid(True, linestyle="--", alpha=0.7)


# Plot for Energy
plot_barbell(axs[0], params_energy, "Peak Location (eV)")
# add twin axis to show the difference in number of standard deviations from the mean
ax_en = axs[0].twinx()
ax_en.plot(
    peak_indices,
    energy_difference_std,
    color="k",
    linestyle="--",
    label="Energy Difference (std)",
    alpha=0.7,
    zorder=0,
)
ax_en.set_ylabel("Difference (std)", fontsize=12, color="k")
ax_en.tick_params(axis="y", labelcolor="k")
ax_en.grid(False)  # Disable grid for twin axis
# Plot for Width
plot_barbell(axs[1], params_width, "FWHM (eV)")
ax_wid = axs[1].twinx()
ax_wid.plot(
    peak_indices,
    width_difference_std,
    color="k",
    linestyle="--",
    label="Width Difference (std)",
    alpha=0.7,
    zorder=0,
)
ax_wid.set_ylabel("Difference (std)", fontsize=12, color="k")
ax_wid.tick_params(axis="y", labelcolor="k")
ax_wid.grid(False)  # Disable grid for twin axis

# Plot for Amplitude
plot_barbell(axs[2], params_amplitude, "Amplitude (a.u.)")
ax_amp = axs[2].twinx()
ax_amp.plot(
    peak_indices,
    amplitude_difference_std,
    color="k",
    linestyle="--",
    label="Amplitude Difference (std)",
    alpha=0.7,
    zorder=0,
)
ax_amp.set_ylabel("Difference (std)", fontsize=12, color="k")
ax_amp.tick_params(axis="y", labelcolor="k")
ax_amp.grid(False)  # Disable grid for twin axis


# Plot for Theta
plot_barbell(axs[3], params_theta, r"Dipole Angle ($\circ$)")
axs[3].set_xlabel("Peak Number", fontsize=12)
ax_theta = axs[3].twinx()
ax_theta.plot(
    peak_indices,
    theta_difference_std,
    color="k",
    linestyle="--",
    label="Theta Difference (std)",
    alpha=0.7,
    zorder=0,
)
ax_theta.set_ylabel("Difference (abs)", fontsize=12, color="k")
ax_theta.tick_params(axis="y", labelcolor="k")
ax_theta.grid(False)  # Disable grid for twin axis
# add white background to the legend
# add the difference to the legend handle list
handles, labels = axs[0].get_legend_handles_labels()
handles.append(ax_en.get_lines()[0])
labels.append("Difference (abs)")
axs[0].legend(
    handles=handles,
    labels=labels,
    loc="upper center",
    fontsize=10,
    frameon=True,
    fancybox=False,
    facecolor="white",
    edgecolor="black",
)
fig.align_labels()
plt.tight_layout()
plt.show()
```

```python
# --- Plotting Setup ---
sns.set_context("paper")
plt.rcParams.update(
    {
        "font.size": 8,
        "axes.labelsize": 8,
        "xtick.labelsize": 8,
        "ytick.labelsize": 8,
        "xtick.direction": "in",
        "ytick.direction": "in",
        "axes.linewidth": 1.0,
        "xtick.major.width": 1.0,
        "ytick.major.width": 1.0,
    }
)

initial_params = {
    "energy": params_energy.loc["pwi0Fit"],
    "fwhm": params_width.loc["pwi0Fit"],
    "amplitude": params_amplitude.loc["pwi0Fit"],
    "theta": params_theta.loc["pwi0Fit"],
}
final_params = {
    "energy": params_energy.loc["pwFinal"],
    "fwhm": params_width.loc["pwFinal"],
    "amplitude": params_amplitude.loc["pwFinal"],
    "theta": params_theta.loc["pwFinal"],
}
std_change = {
    "energy": energy_difference_std,
    "fwhm": width_difference_std,
    "amplitude": amplitude_difference_std,
    "theta": theta_difference,
}

# Define colors for clarity and aesthetic appeal
initial_color = "dimgray"
# Use a color palette for the "refined" points based on significance
significance_palette = sns.color_palette("vlag_r", n_colors=5)  # A divergent palette
bar_color = "lightgray"


# --- Main Plotting Function ---
def plot_parameter_change(
    ax,
    x_indices,
    initial_vals,
    final_vals,
    std_vals,
    ylabel,
    significance_threshold=1.0,
):
    """
    Creates a dumbbell plot where the final marker color indicates the
    significance of the change.
    """
    # Determine colors for final points based on significance
    final_colors = [significance_palette[2]] * len(
        x_indices
    )  # Default to center color (no change)
    for i, std in enumerate(std_vals):
        if std > significance_threshold:
            final_colors[i] = significance_palette[0]  # Positive significant change
        elif std < -significance_threshold:
            final_colors[i] = significance_palette[4]  # Negative significant change

    # Plot vertical lines (stems of the dumbbell)
    ax.vlines(
        x_indices,
        initial_vals,
        final_vals,
        color=bar_color,
        linestyles="-",
        linewidth=1.0,
        zorder=1,
    )

    # Plot markers for initial and final values
    ax.scatter(
        x_indices, initial_vals, color=initial_color, s=10, zorder=2, label="Initial"
    )
    # Plot final values point by point to assign correct colors
    for i in range(len(x_indices)):
        ax.scatter(x_indices[i], final_vals[i], color=final_colors[i], s=12, zorder=3)

    ax.set_ylabel(ylabel)
    return ax


# --- Figure Creation ---
fig, axs = plt.subplots(
    4, 1, figsize=(3.0, 4.5), sharex=True, dpi=300, gridspec_kw={"hspace": 0.1}
)

# Plot for each parameter
plot_parameter_change(
    axs[0],
    peak_indices,
    initial_params["energy"],
    final_params["energy"],
    std_change["energy"],
    "Energy (eV)",
)
plot_parameter_change(
    axs[1],
    peak_indices,
    initial_params["fwhm"],
    final_params["fwhm"],
    std_change["fwhm"],
    "FWHM (eV)",
)
plot_parameter_change(
    axs[2],
    peak_indices,
    initial_params["amplitude"],
    final_params["amplitude"],
    std_change["amplitude"],
    "Amplitude (a.u.)",
)
plot_parameter_change(
    axs[3],
    peak_indices,
    initial_params["theta"],
    final_params["theta"],
    std_change["theta"],
    "Theta ()",
)

# --- Final Touches and Customization ---
for ax in axs:
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.yaxis.set_major_locator(plt.MaxNLocator(nbins=4, prune="both"))
    ax.yaxis.set_minor_locator(plt.NullLocator())

# Customize the bottom axis
axs[3].set_xlabel("Peak Number")
axs[3].xaxis.set_major_locator(plt.MultipleLocator(4))
axs[3].xaxis.set_minor_locator(plt.MultipleLocator(1))
axs[3].tick_params(axis="x", which="major", length=4)
axs[3].tick_params(axis="x", which="minor", length=2)

# Create a clear, unified legend for the whole figure
# Add dummy points for the legend
legend_elements = [
    plt.Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label="Initial",
        markerfacecolor=initial_color,
        markersize=5,
    ),
    plt.Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label=f"Refined (> {1})",
        markerfacecolor=significance_palette[0],
        markersize=5,
    ),
    plt.Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label="Refined (|| < 1)",
        markerfacecolor=significance_palette[2],
        markersize=5,
    ),
    plt.Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label=f"Refined (< -{1})",
        markerfacecolor=significance_palette[4],
        markersize=5,
    ),
]
fig.legend(
    handles=legend_elements,
    loc="upper center",
    bbox_to_anchor=(0.5, 1.02),
    ncol=4,
    frameon=False,
    fontsize=7,
)

fig.align_labels()
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make space for the legend
plt.show()
```

```python
import matplotlib.patheffects as path_effects

plt.style.use("seaborn-v0_8-whitegrid")  # A clean base style
fig, axes = plt.subplots(4, 1, figsize=(8, 10), sharex=True)
plt.subplots_adjust(hspace=0.1)  # Reduce space between plots

# Define colors and the glow effect
increase_color = "#d9534f"  # A bootstrap-style red
decrease_color = "#428bca"  # A bootstrap-style blue
initial_color = "#555555"
glow_color = "#FFE792"  # A light, buttery yellow
magic_angle_color = "#FFD700"  # Gold for the star

# The glow effect will be a "shadow" drawn behind the main element
glow_effect = None  # [path_effects.SimpleLineShadow(offset=(0, 0), shadow_color=glow_color, alpha=0.8, linewidth=5)]
peak_numbers = np.arange(1, 18)  # Peak numbers from 1 to 17
num_peaks = len(peak_numbers)

# --- 3. Panel 1: Change in Energy (Lollipop Plot) ---
ax1 = axes[0]
delta_energy = energy_difference_std
colors_energy = [increase_color if d > 0 else decrease_color for d in delta_energy]

# Plot vertical lines (stems)
ax1.vlines(
    peak_numbers,
    0,
    delta_energy,
    color=colors_energy,
)  # path_effects=glow_effect, lw=2)
# Plot dots on top
ax1.scatter(
    peak_numbers, delta_energy, color=colors_energy, s=50, zorder=3
)  # , path_effects=glow_effect)

ax1.axhline(0, color="gray", linestyle="--", linewidth=1)
ax1.set_ylabel(" Energy (std.)", fontsize=12, weight="bold")
ax1.tick_params(axis="y", labelsize=11)
ax1.grid(axis="x")  # Keep vertical grid lines only

# --- 4. Panel 2: Change in FWHM (Lollipop Plot) ---
ax2 = axes[1]
delta_fwhm = width_difference_std
colors_fwhm = [increase_color if d > 0 else decrease_color for d in delta_fwhm]

ax2.vlines(
    peak_numbers,
    0,
    delta_fwhm,
    color=colors_fwhm,
)  # path_effects=glow_effect, lw=2)
ax2.scatter(
    peak_numbers, delta_fwhm, color=colors_fwhm, s=50, zorder=3
)  # , path_effects=glow_effect)

ax2.axhline(0, color="gray", linestyle="--", linewidth=1)
ax2.set_ylabel(" FWHM (std.)", fontsize=12, weight="bold")
ax2.tick_params(axis="y", labelsize=11)
ax2.grid(axis="x")

# --- 5. Panel 3: Amplitude Refinement (Barbell Plot) ---
ax3 = axes[2]
for i in range(num_peaks):
    y_initial = initial_params["amplitude"][i]
    y_refined = final_params["amplitude"][i]
    color = increase_color if y_refined > y_initial else decrease_color

    # Barbell line
    ax3.plot(
        [peak_numbers[i], peak_numbers[i]],
        [y_initial, y_refined],
        c=color,
        lw=2,
        zorder=1,
        path_effects=glow_effect,
    )
    # Initial point (no glow)
    ax3.scatter(peak_numbers[i], y_initial, c=initial_color, s=40, zorder=2)
    # Refined point
    ax3.scatter(
        peak_numbers[i], y_refined, c=color, s=60, zorder=3, path_effects=glow_effect
    )

ax3.set_ylabel("Amplitude (a.u.)", fontsize=12, weight="bold")
ax3.tick_params(axis="y", labelsize=11)
ax3.grid(axis="x")

# --- 6. Panel 4: Theta Refinement (Barbell Plot with Highlighting) ---
ax4 = axes[3]
ax4.axhline(55, color="gray", linestyle=":", linewidth=1.5, label="Magic Angle (55)")

for i in range(num_peaks):
    y_initial = initial_params["theta"][i]
    y_refined = final_params["theta"][i]
    color = increase_color if y_refined > y_initial else decrease_color

    # Barbell line
    ax4.plot(
        [peak_numbers[i], peak_numbers[i]],
        [y_initial, y_refined],
        c=color,
        lw=2,
        zorder=1,
        path_effects=glow_effect,
    )
    # Initial point (no glow)
    ax4.scatter(peak_numbers[i], y_initial, c=initial_color, s=40, zorder=2)
    # Refined point
    ax4.scatter(
        peak_numbers[i], y_refined, c=color, s=60, zorder=3, path_effects=glow_effect
    )

    # Check for magic angle crossover and add star
    if y_initial < 55 and y_refined > 55 or y_initial > 55 and y_refined < 55:
        ax4.scatter(
            peak_numbers[i],
            y_refined,
            c=magic_angle_color,
            s=300,
            marker="*",
            zorder=4,
            label="Magic Angle Crossover",
        )


ax4.set_ylabel("Theta ()", fontsize=12, weight="bold")
ax4.tick_params(axis="y", labelsize=11)
ax4.grid(axis="x")

# --- 7. Final Touches and Legend ---
ax4.set_xlabel("Peak Number", fontsize=12, weight="bold")
ax4.set_xticks(peak_numbers)
ax4.tick_params(axis="x", labelsize=11)

# Clean up spines for a minimalist look
for ax in axes:
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_linewidth(1.2)
    ax.spines["bottom"].set_linewidth(1.2)

# Create a custom legend
from matplotlib.lines import Line2D

legend_elements = [
    Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label="Parameter Increased",
        markerfacecolor=increase_color,
        markersize=10,
    ),
    Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label="Parameter Decreased",
        markerfacecolor=decrease_color,
        markersize=10,
    ),
    Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label="Initial Value",
        markerfacecolor=initial_color,
        markersize=9,
    ),
    Line2D(
        [0],
        [0],
        marker="*",
        color="w",
        label="Magic Angle Crossover",
        markerfacecolor=magic_angle_color,
        markersize=15,
    ),
]
fig.legend(
    handles=legend_elements,
    loc="upper center",
    bbox_to_anchor=(0.5, 1.02),
    ncol=4,
    fontsize=11,
)
fig.align_labels()
plt.tight_layout()
plt.show()
```
